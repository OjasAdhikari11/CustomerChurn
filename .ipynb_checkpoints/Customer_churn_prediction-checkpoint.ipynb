## Creating Cohort based on Tenure

```python
df.groupby(['Churn','tenure']).count().transpose()
```

```python
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
```

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
```

```python
churn_rate = 100*yes_churn/(no_churn+yes_churn)
```

```python
churn_rate.transpose()['customerID']
```

```python
churn_rate.transpose()['customerID'].plot()
```

## Broader Cohort Groups

```python
def cohort(tenure):
    if tenure<13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure <49:
        return '24-48 Months'
    else:
        return "over 48 Months"
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['Tenure Cohort','tenure']]
```

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.scatterplot(data = df, x= 'MonthlyCharges', y = 'TotalCharges',hue = 'Tenure Cohort',alpha = 0.4,linewidth = 0.3,palette = 'Dark2')
```

```python
sns.countplot(data =df,x = 'Tenure Cohort', hue = 'Churn')
```

## Single Decision Tree

```python
X = df.drop(['Churn','customerID'],axis = 1)
```

```python
X = pd.get_dummies(X,drop_first = True)
```

```python
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth = 6)
```

```python
dt.fit(X_train,y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix,classification_report
```

```python
preds = dt.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
plot_confusion_matrix(dt,X_test,y_test)
```

```python
imp_feats = pd.DataFrame(data = dt.feature_importances_,index = X.columns,columns = ['Feat Imp'])
```

```python
imp_feats = imp_feats.sort_values('Feat Imp')
```

```python
imp_feats
```

```python
imp_feats = imp_feats[imp_feats['Feat Imp']>0]
```

```python
imp_feats
```

```python
plt.figure(figsize= (14,6),dpi = 200)
sns.barplot(data = imp_feats, x = imp_feats.index, y = 'Feat Imp')
plt.xticks(rotation  = 90);
```
